```python
tsla = yf.Ticker("TSLA")

df = tsla.history(period="max")

df.head()
```

```python
revenue_df = tsla.financials.loc['Total Revenue']

revenue_df = revenue_df.to_frame().T

revenue_df.tail()
```

```python
gme = yf.Ticker("GME")

gme_df = gme.history(period="max")

gme_df.head()
```

```python
revenue_df = gme.financials.loc['Total Revenue']

revenue_df = revenue_df.to_frame().T

revenue_df.tail()
```

```python
def make_graph(stock_data, title):
    plt.figure(figsize=(12,6))
    plt.plot(stock_data.index, stock_data['Close'], color='green')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.grid(True)
    plt.show()

def make_graph_tsla_revenue(tsla):
    revenue = tsla.financials.loc['Total Revenue']
    revenue = revenue.sort_index()

    plt.figure(figsize=(12,6))
    plt.plot(revenue.index, revenue.values, marker='o', color='green')
    plt.title("Tesla (TSLA) Total Revenue Over Time")
    plt.xlabel("Year")
    plt.ylabel("Revenue (USD)")
    plt.grid(True)
    plt.show()
```

```python
make_graph(df, "Tesla Total Price Action")

make_graph_tsla_revenue(tsla)
```

```python
def make_graph_gme(gme_data):
    plt.figure(figsize=(12,6))
    plt.plot(gme_data.index, gme_data['Close'], color='purple')
    plt.title("GameStop Total Price Action")
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.grid(True)
    plt.show()

def make_graph_gme_revenue(gme):
    revenue = gme.financials.loc['Total Revenue']
    revenue = revenue.sort_index()

    plt.figure(figsize=(12,6))
    plt.plot(revenue.index, revenue.values, marker='o', color='orange')
    plt.title("GameStop (GME) Total Revenue Over Time")
    plt.xlabel("Year")
    plt.ylabel("Revenue (USD)")
    plt.grid(True)
    plt.show()
```

```python
make_graph_gme(gme_df)

make_graph_gme_revenue(gme)
```
